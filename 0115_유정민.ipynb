## duplicated()
 잘못된 열 수정하고 중복 처리 우선
  - PART 1
    * duplicated() 최초값 빼고 나머지 중복만 나온다.
      - [주로 중복데이터 갯수]
    * duplicated(keep = False) 최초값 포함한 모든 중복이 나온다.
      - [중복데이터끼리 묶어서 출력 후 확인용]

  - PART 2 반전 (반전할 때는 처리할 때 사용)
    * ~duplicated() 최초값이랑 유일값만 남긴다.( 중복 x )
      -[주로 처리용]
    * ~duplicated(keep=False) 원래부터 1개 였던 (유일값)만 남긴다.
      

```python
# 최초의 값 뺴고 나머지 중복 (불리언 값)
df_sample = pd.DataFrame({
        'id':['a','a','b','c','b'],
        'age':[20, 30, 40, 50, 50]
})
df_sample

# 중복값을 조절
```
