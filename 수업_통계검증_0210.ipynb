### 상관관계 검증 spearman
(귀무가설)H0: 매니저의 소통과 직업 만족도는 상관관계를 보이지 않는다.
(대립가설)H1: 매니저의 소통과 직업 만족도는 상관관계를 보인다.
  * 한쪽이 늘어날 때 다른 한쪽이 늘어나거나 줄어들면 "통계적 관점"에서 다른 한쪽이 늘어나거나(양) 줄어든다(음).

# ** [3. 집단 차이 검증]
- 정규분포 여부에 따라 모수,비모수 선택

## 집단 간 평균차이
- 두 집단간 평균(연속형변수의) 이 통계적으로 유의미하게 차이가 나는가?
- 예) 과장과 차장 직급간 연봉차이가 유의미하게 날까?

## 검정방법 선택
1. 두 집단의 차이만 검증하면 되는가? (yes -> 2,     No -> 3)
2. 두 집단 모두 정규분포를 보이는가? (yes -> t검정, No -> Welch_anova)
3. 여러집단 모두가 정규분포를 보이는가? (yes -> anova, No -> Welch_anova)


## T- test
- 과장, 차장 두 직급의 연봉 평균 차이
- 귀무가설(H0): 과장과 차장의 연봉의 평균차이가 유의미 하지 않다.
- ~대립가설(H1): 과장과 차장의 연봉의 평균차이가 유의미 하다.~
- 현재 표본과 유의수준을 기준으로 통계적으로 유의미하지 않다.
- 표본의 크기가 너무 작으면 실제 존재하는 차이를 검출하지 못할 수 있다.
- **의사결정시**
  * 추가적인 맥락을 고려하세여! (사내 특이사항, 구조, 부서 특수성)
- 데이터 분석에서 **T-검정(T-test)**이란 **"두 집단의 평균값이 통계적으로 정말 차이가 나는지"**를 확인하는 방법입니다.

```python
# 정규성 검증
s, p = shapiro(df[df['직급'] == '과장']['연봉(천만원)'])
print(f'과장: {p}')
s, p = shapiro(df[df['직급'] == '차장']['연봉(천만원)'])
print(f'차장: {p}')
# 둘다 정규분포를 따라서 모수 검증 가능
```

```python
# 독립표본 t
과장그룹 = df[df['직급']== '과장']['연봉(천만원)']
차장그룹 = df[df['직급']== '차장']['연봉(천만원)']
t, p = ttest_ind(과장그룹,차장그룹)
if p < 0.05:
  print("과장과 차장의 연봉의 평균차이가 유의미하다.")
else:
  print("과장과 차장의 연봉의 평균차이가 유의미하지않다.")
# 차이가 없는 것이 아님
```

## 집단 간 중앙값 차이

- 귀무가설: 중분류별 알라딘판매점수 중앙값의 차이가 유의미하지 않다.

```python
# kruskal 검증
group = [group for name, group in book_ver1.groupby('공식중분류')['알라딘_판매점수']]
group
```

```python
# 비모수
from scipy.stats import kruskal

# Perform the Kruskal-Wallis H-test
h_statistic, p_value = kruskal(*group)

print(f"Kruskal-Wallis H-statistic: {h_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("5% 유의수준에서 귀무가설을 기각합니다. 중분류별 알라딘 판매점수 중앙값의 차이가 유의미합니다.")
else:
    print("5% 유의수준에서 귀무가설을 기각할 수 없습니다. 중분류별 알라딘 판매점수 중앙값의 차이가 유의미하다고 볼 수 없습니다.")
```

### 관점에 따른 분석 story

#### 중앙값 관점
- '독일문학'과 '한국문학'의 중앙값 차이는 "대부분(전형적인)" 해당분류도서의

- '독일문학'에 일부 높은 판매점수를 가진 도서가 있어도 중앙값은 흔들리지 X
- 따라서 전형적으로 (시작점부터 끝까지) 높거나 해서 중앙값 자체가 다른 수준으로 잡히는 등의 전형적인 판매점수의 차이

- **중앙값(Median)**이 가진 통계적 특성과 이를 데이터 해석에 어떻게 활용하는지를 설명하고 있습니다. 핵심은 '평균(Mean)의 함정'에 빠지지 않고 집단의 실제 수준을 파악하는 방법에 대한 것입니다.

- 중앙값 1 등부터 꼴지의 줄을 세운 뒤 정확히 중간에 서있는 사람의 점수 (슈퍼스타의 존재와 무관)

-즉, 중앙값을 본다는 것은 소수의 대박 상품에 현혹되지 않고, 해당 카테고리 도서들이 보통 어느 정도 수준에서 팔리는지를 냉정하게 보겠다는 의지입니다.

(한 집단은 다른 집단과 비교했을 때 값의 전반적인 위치가 다르다.)

#### 평균 관점
- 일부 극단적으로 높은 판매점수를 가진 도서가 얼마나 섞여있느냐
- 그럼에도 불구하고, 평균에 대한 강건한 추론이 필요할 때
