## 신뢰수준에 따른 표본오차 (예 ±3.1%)
- z분포 임계값
  * 모분산을 알거나 표본이 충분이 큰 경우
  - 수준95% : 1.96
  - 수준99% : 2.576
  - 수준90% : 1.645
- t분포 임계값
 * 표본으로 부터 직접 임계값을 구함

 **임계값(Critical Value)**이란, 가설 검정 시 귀무가설을 기각할지 여부를 결정하는 **최종적인 경계선(Boundary)**을 의미합니다.


```python
import pandas as pd

df = pd.read_csv('employee_data_example.csv')
df.head()
```

```python
# z분포로 구한 것 - 모집단이 있거나 표본이 충분히 큰 경우 < 하지만 기준은 없음
import numpy as np
## 연봉의 신뢰구간 ##
# 평균
mean_salary = df['연봉(천만원)'].mean()
# 표준편차
std_salary = df['연봉(천만원)'].std(ddof=1) # n-1 표준편차 (모집단의 표준편차를 구할땐 ddof=0)
# 샘플크기
n = len(df['연봉(천만원)'])

## 임계값
# z-분포  (갯수충분)
z_critical = 1.960

# # t-분포 (갯수 불충분)
# c_level = 0.95 # 95% 신뢰수준
# alpha = 1 - c_level
# t_critical = t.ppf(1 - alpha/2, df= n-1)

# 오차 한계
margin_of_error = z_critical * (std_salary / np.sqrt(n))
#                 t_critical (t임계값일 경우)

# 신뢰구간 -> (평균 - 오차한계) ~ (평균 + 오차한계)
lower_bound = mean_salary - margin_of_error
upper_bound = mean_salary + margin_of_error

print(f"신뢰구간: ({lower_bound}, {upper_bound})")
```

**표본 오차(Sampling Error)**란 모집단 전체를 조사하지 않고 그 일부인 **표본(Sample)**만을 추출하여 조사했을 때 발생하는 통계적 수치와 실제 모집단 값(모수) 사이의 차이를 의미

```python
df['연봉(천만원)'].mean()
```

```python
print(f'오차:{margin_of_error}')
print(f'신뢰구간 오차는 평균대비 ±{margin_of_error/df['연봉(천만원)'].mean()}%')
print(f"신뢰구간: ({lower_bound}, {upper_bound})")
```

```python
#프로젝트에서 그대로 사용->
print(f'95%  신뢰수준에서 해당 표본의 평균연봉 신뢰구간은 "신뢰구간: ({lower_bound}, {upper_bound})"')
print(f'95% 신뢰수준에서 위 구간은 모수를 포함 할 것으로 신뢰되는 구간이다')
```

# 1. 정규성 검증(정규분포인지 아닌지를 검증)
- 모수, 비모수 검정(test)방법 결정
- 일반적인 귀무가설은 '차이가 없다 or 관계가 없다' 인데 반해
- 정규성/동질성 검증은 '변화가 없다'로 설정
- 즉 H0:"정규분포를 보인다"

## 이유
- 많은 집단들은(분야를 불문하고) 실제로 정규분포를 보입니다.
- 만약, 귀무가설이 [정규분포를 보이지 않는다] 였다면, 표본 데이터상의 작은(일부의) 노이즈로 인해 실제론 정규분포를 띄지만 검증에 의해 정규분포가 아니라는(이때, p > 0.05) 판단오류를 내릴 확률이 있습니다.
- "정규분포가 아니다" 라는 확실한 증거를 찾아야지만 귀무가설을 기각하도록(안전장치)만들기 위해, 귀무가설을 "데이터는 정규분포를 보인다" 로 설정.

- 따라서, "관계를 보이지 않는다"가 보수적인 선택인 것 처럼, "정규분포이다" 가 보수적인 선택.
- **"관계를 보이지 않는다"(보수적)**: "왠만해선 관계를 보이지 않지만, 확실하게 관계를 보임을 주장하기 위해 증거를 끌어모아 귀무가설을 기각하겠다."
- **"정규분포이다"(보수적)**: "대부분 정규분포이지만, 확실하게 정규분포가 아님을 입증하려면 증거를 끌어모아 귀무가설을 기각해 봐라."

## 결론
  - 귀무가설(H0):정규분포다 (p > 0.05)
  - 대립가설(H1): 정규분포가 아니다 (p < 0.05)
## 검정방법
- shapiro-wilk
- QQ plot

### shapiro-wilk test
  - 예를 들어 직급별 연봉 평균차이를 검증하고 싶다.
  - 직급별 연봉이 정규분포를 보이는지를 먼저보고 (shapiro-wilk test)
  - 평균차이를 검증하는 방법을 선택하고자 한다.(t-test, anova, welch-anova)

```python
# 선제조건 - 정규성검증이 필요

df[df['직급']=='차장']['연봉(천만원)']

from scipy.stats import *

s, p = shapiro(df[df['직급']=='차장']['연봉(천만원)'])
print(f'p-value: {p}')
```

## 결론
  - 귀무가설(H0):차장의 연봉은 정규분포다 (p > 0.05)
  - 대립가설(H1): 차장의 연봉은 정규분포가 아니다 (p < 0.05)

귀무가설을 기각할 근거가 충분하지 않다


### QQ plot

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
sm.qqplot(df[df['직급'] == '차장']['연봉(천만원)'], line='s')
plt.show()
# 가이드라인에 가까우면 가까울수록 정규분포임을 의미합니다.
# 어느정도 모인형태가 보인다면 정규분포라고 봐도 괜찮다.
```
