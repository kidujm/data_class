```python
key_cols = ["기준_연월", "행정동_코드"]
if "시간대_코드" in 경기도df.columns:
    key_cols.append("시간대_코드")

print("키 컬럼:", key_cols)
print("키 기준 중복 행 수:", 경기도df.duplicated(subset=key_cols).sum())
```

```python
# 타입정리
```

```python
# 기준_연월: YYYYMM 형태로 정리
경기도df["기준_연월"] = (
    경기도df["기준_연월"]
    .astype(str)
    .str.replace(r"\.0$", "", regex=True)
    .astype(int)
)

# 날짜 컬럼(월 기준) 생성
경기도df["기준_월"] = pd.to_datetime(
    경기도df["기준_연월"].astype(str) + "01",
    format="%Y%m%d"
)

# 연도 / 월 파생
경기도df["연도"] = 경기도df["기준_월"].dt.year
경기도df["월"] = 경기도df["기준_월"].dt.month
```

```python
for col in ["행정동_코드", "시간대_코드"]:
    경기도df[col] = (
        경기도df[col]
        .astype(str)
        .str.replace(r"\.0$", "", regex=True)
        .str.strip()
    )
```

```python
num_cols = [
    "유동인구_수",
    "유동인구_비율",
    "전월_대비_증감_인구수",
    "전월_대비_증감률",
    "전년_동월_대비_증감_인구수",
    "전년_동월_대비_증감률",
]

for col in num_cols:
    경기도df[col] = pd.to_numeric(
        경기도df[col].astype(str).str.replace(",", ""),
        errors="coerce"
    )
```
