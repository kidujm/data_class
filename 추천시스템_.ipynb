```python
# 영화 데이터
from surprise import Dataset
import pandas as pd

data = Dataset.load_builtin('ml-100k')
df = pd.DataFrame(data.raw_ratings)
df
```

```python
## 필요없는 열 미리 제거
df.drop([3], axis=1, inplace=True)
df
```

```python
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

from surprise import accuracy
from surprise import SVD
# 학습 및 검증
# 알고리즘 선택 및 학습 (SVD 사용)
algo = SVD()
algo.fit(trainset)

# 테스트셋에 대한 예측 및 성능 평가
pred = algo.test(testset) # 예측
rmse = accuracy.rmse(pred) # 평가
```

```python
def recommend_movies(user_id):
    # 전체 아이템(영화) 목록
    all_items = set(df[1].unique())

    # 유저가 이미 평점을 준 아이템 목록
    user_items = set(df[df[0] == user_id][1])

    # 평점을 주지 않은 아이템 목록
    unrated = all_items - user_items

    # 예측 수행
    predictions = [algo.predict(user_id, iid) for iid in unrated]
    predictions = pd.DataFrame(predictions)

    # 추천 점수 테이블
    추천 = predictions[['iid', 'est']]

    # 점수 기준 정렬
    정렬 = 추천.sort_values('est', ascending=False).reset_index(drop=True)

    # 영화 제목 매핑 테이블
    movies.rename(columns={0: 'iid', 1: 'title'}, inplace=True)
    맵핑테이블 = movies.groupby('iid')['title'].first()

    # iid 타입 맞추기
    정렬['iid'] = 정렬['iid'].astype(int)

    # 제목 매핑
    정렬['title'] = 정렬['iid'].map(맵핑테이블)

    # 상위 10개 반환
    return 정렬.head(10)
```
