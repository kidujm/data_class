```python

import requests
import pandas as pd
from bs4 import BeautifulSoup


def weather(지역명):
    url = f'https://search.naver.com/search.naver?query={지역명}%EB%82%A0%EC%94%A8'
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')

    # 현재 온도
    온도 = soup.find('div', class_='temperature_text').text
    온도 = float(온도.replace('현재 온도', '').strip()[:-1])

    # 날씨 상태
    날씨상태 = soup.find('span', class_='weather before_slash').text

    

    # 체감온도 / 습도
    summary = soup.find('dl', class_='summary_list')
    info = {}

    for div in summary.find_all('div', class_='sort'):
        k = div.find('dt').text.strip()
        v = div.find('dd').text.strip()
        v = v.replace('°', '').replace('%', '').replace('m/s', '')
        info[k] = float(v)

    # DataFrame 생성
    df = pd.DataFrame([{
        '지역': 지역명,
        '온도': 온도,
        '날씨상태': 날씨상태,
        '체감': info.get('체감'),
        '습도': info.get('습도')
    }])

    return df

print()


 
```
