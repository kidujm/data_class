### **굳이** 스크롤 내려야 함   

```python
# n번만 스크롤 내리기
def scroll_n_times(driver, n, interval=0.02):
    body = driver.find_element("tag name", "body")
    for _ in range(n):
        body.send_keys("\ue00f")  # PAGE_DOWN
        time.sleep(interval)
# 사용 예시

scroll_n_times(driver, n=30)
```

# 명령
- 데이터프레임 book에 d_url이 실제 각 페이지의 주소이다.
- 남성에서 가장 높은 %를 보이는 연령대, 여성에서 가장 높은 %를 보이는 연령대를 추출해서 '남성_top_연령대', '여성_top_연령대' 에 추가.
- 첫 3개 페이지만 테스트 해보고 싶다.
- 패키지로드와 드라이버 세팅 코드 필요없음.

```python
from bs4 import BeautifulSoup

male_top_ages = []
female_top_ages = []

for idx, url in enumerate(book['d_url'][:3]):
    driver.get(url)
    scroll_n_times(driver, n=30)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    buyer_div = soup.find('div', 'buyer_area_a')

    male_max = {'age': None, 'per': -1}
    female_max = {'age': None, 'per': -1}

    rows = buyer_div.find_all('tr')

    for row in rows:
        age_td = row.find('td', class_='tb_tit')
        if not age_td:
            continue

        age = age_td.get_text(strip=True)

        female_td = row.find('td', class_='tb_woman')
        male_td = row.find('td', class_='tb_man')

        if female_td:
            female_per = female_td.find('div', class_='per')
            if female_per:
                val = float(female_per.text.replace('%', ''))
                if val > female_max['per']:
                    female_max = {'age': age, 'per': val}

        if male_td:
            male_per = male_td.find('div', class_='per')
            if male_per:
                val = float(male_per.text.replace('%', ''))
                if val > male_max['per']:
                    male_max = {'age': age, 'per': val}

    male_top_ages.append(male_max['age'])
    female_top_ages.append(female_max['age'])

book.loc[:2, '남성_top_연령대'] = male_top_ages
book.loc[:2, '여성_top_연령대'] = female_top_ages
```

```python
result_df = pd.DataFrame({
    'd_url': book.loc[:2, 'd_url'].values,
    '남성_top_연령대': male_top_ages,
    '여성_top_연령대': female_top_ages
})

result_df
```
