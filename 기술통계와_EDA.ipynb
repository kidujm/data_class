```python
book['최다연령대_여성'].value_counts()
```

- **시각차트용 코드는 챗봇에게 요청**
- 범주의 갯수가 적을 때는 파이차트


```python
import pandas as pd
import matplotlib.pyplot as plt

# 예시 데이터프레임 (실제 데이터프레임에서 book 변수로 지정되었다고 가정)
# book = pd.read_csv('your_data.csv')  # 실제 데이터 불러오기 예시

# value_counts를 사용해 값의 분포 계산
value_counts = book['최다연령대_여성'].value_counts()

# 파이차트 시각화
plt.figure(figsize=(8, 8))
plt.pie(
    value_counts,
    labels=value_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False
)
plt.title('최다연령대 여성 비율')
plt.axis('equal')  # 원형 유지
plt.show()
```

- 두 성별 모두 40%의 비중이 가장 높음.
- 남성의 타깃(20-30) 비중 27%
- 여성의 타깃(20-30) 비중 36%
- 단, 여성 타깃 연령대에서 높은 인기를  끌고 있는 (20~30대_여성 IQR 75% 구간이상) 그룹의 도서가 데이터 내에 존재.

## Percentiles
- 직접 구간(%)을 지정해 describe 출력물을 확인

```python
book.describe(percentiles=[0.3,0.6,0.9])[['publication_year','loan_count', 'sales_point','price','page_count', 'review_count', 'average_rating']]
```

```python
book.describe(percentiles=[0.9,0.92,0.94,0.96,0.98])[['publication_year','loan_count', 'sales_point','price','page_count', 'review_count', 'average_rating']]
```

- 노이즈( 잘못된 정보냐 ) / 이상치 ( 극단 값이냐 )
- 평균이 전체적인 대푯값이라는건 틀림 (평균이 왜곡됨)
- 극단치가 분석내용과 일치할 때 평균 값을 사용
- 이상치가 있음을 알고 있음에도 중요한 정보면 평균 값을 사용
- 중앙값이 대표값

# 중심경향성
- 데이터가 어떤값을 중심으로 분포? - 어디에 몰려있나?

## 평균
- 모든 값을 더한 후 데이터 갯수로 나눈 값
- '경향'을 반영하나, '이상치'에 민감. (왜곡 가능성이 높다.)
  * 이상치 : 극단적으로 높거나 낮은 값 (sales-point 70만이 넘는 숫자)

```python
book['loan_count'].mean()
```

- "대출건수 평균 '714.19' 우리 표본을 대표하는 대출건수인가??"
  * 중심경향성을 대표하는 대표값에 대한 질문
- 대출건수의 평균은 극단적으로 높은 이상치에 의해 왜곡되었다고 판ㄴ단되어 대표값은 중앙값인 560.5으로 설정
- 따라서, 2025년 정보나루에서 수집된 1년간의 대출건수 기준 랭킹 데이터 표본의 '대출건수'는 1권당 약 560.5건을 기록했습니다.

## *중앙값*
- 숫자
- 데이터를 정렬했을 때 한 가운데 위치한 값
- IQR 50% 구간의 값
- 이상치의 영향을 평균보다 훨씬 덜 받습니다.
- 일반적인 경우 이상치(극단값)이 존재할 때 중앙값을 대표값을 활용합니다.

```python
book['loan_count'].median()
```

- 중심경향성/ 분포 / 히스토그램

```python
# 대출건수와 sales_point 히스토그램
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(book['loan_count'], bins=20, color='skyblue', edgecolor='black')
plt.title('대출 건수 분포 (Loan Count Distribution)')
plt.xlabel('대출 건수 (Loan Count)')
plt.ylabel('도서 수 (Frequency)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(book['sales_point'], bins=20, color='skyblue', edgecolor='black')
plt.title('판매 포인트 분포 (Sales Point Distribution)')
plt.xlabel('판매 포인트 (Sales Point)')
plt.ylabel('도서 수 (Frequency)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
