##  알고리즘 결정트리(DecisionTreeClassifier)
- 여러가지 알고리즘을 모아놓은 것이 사이킷런

```python
# 패키지 로드
from sklearn.datasets import load_iris # 붓꽃데이터
from sklearn.tree import DecisionTreeClassifier # 결정트리 알고리즘
from sklearn.model_selection import train_test_split # 데이터셋 분리
```

사이킷런의 대표 분류 알고리즘 5가지와 간결한 설명은 다음과 같습니다:

1.  **결정 트리 (Decision Tree)**: 질문을 통해 데이터를 분할하며 분류 규칙을 형성하는, 직관적인 트리 구조 모델입니다.

2.  **서포트 벡터 머신 (Support Vector Machine, SVM)**: 데이터를 가장 잘 나누는 '최적의 결정 경계'를 찾아 분류하는 알고리즘입니다.

3.  **랜덤 포레스트 (Random Forest)**: 여러 개의 결정 트리를 만들어 결과를 종합하여 과적합을 줄이고 예측 성능을 높이는 앙상블 기법입니다.

4.  **K-최근접 이웃 (K-Nearest Neighbors, KNN)**: 새로운 데이터 주변의 가장 가까운 K개 이웃들의 클래스를 참고하여 분류하는 간단한 방식입니다.
거리기반 - 스케일 필수

5.  **로지스틱 회귀 (Logistic Regression)**: 데이터가 특정 클래스에 속할 확률을 예측하여 분류를 수행하는 통계적 모델입니다.

```python
import pandas as pd

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['label'] = iris.target
df_iris
```

```python
iris.target_names
# 데이터프레임으로 만들기
df_species = pd.DataFrame(data=iris.target_names, columns=['species'])
df_species
```

- 머신러닝용 데이터 정제완료

## 데이터셋분리
- 데이터를 학습(80)/ 검증용(20)으로 나눕니다.
- 검증용 데이터의 비율이 일반적으로 20%
- 120(공부)/30(시험)
- train_test_split 패러미터
  * test_size 검증용 데이터 비율(0.2)
  * random_state 시드값. 학습에 일관성 부여.

- x,y -> 피처와 타깃
- x: 피처 :학습용 독립변수
  * X_train : 학습용 피처
  * X_test : 검증용 타깃
  
- y: 타깃(레이블) :정답


```python
df_iris
```

```python
# 피쳐와 레이블 나누기
X = df_iris.drop('label', axis=1)
y = df_iris['label']
```

```python
# 학습용, 검증용 데이터셋 나누기
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=110)
```

# 학습

```python
# 머신러닝 모델로드
dt = DecisionTreeClassifier()

# 학습
dt.fit(X_train, y_train)
```

# 검증
- 검증은 반드시 검증용 데이터(test)를 사용
- 'X_test를 예측한 결과' 와 실제 정답인 'y_test'를 비교해서 채점
-  [predict]             와             [answer]에 각각 위 내용을 저장


```python
dt.predict(X_test)
```

```python
# 모델이 예측한 값을 데이터프레임에 추가
X_test['predict'] = dt.predict(X_test)
```

```python
#기존에 존재하던 답을 answer에 추가
X_test['answer'] = y_test
X_test
```

## 정확도 평가
- predict가 실제 answer과 얼마나 일치 하는지.
- accuracy_score 활용

```python
from sklearn.metrics import accuracy_score
```

```python
#1.
# accuracy_score(y_test,dt.predict(X_test))
#2.
accuracy_score(X_test['answer'],X_test['predict'])
```

## 교차검증
- 학습데이터에 대해서만 성능이 높은 "과적합"
  * 특히, label 분포가 일정하지 않은 경우 ( 품종 0,1,2의 갯수가 제각각인 경우)
- train/test 분리를 여러번 시도 합니다.
- 비율은 동일하지만, train와 train 에 들어가는 데이터 조합이 다름.

- "최종적으로, 알고리즘을 선택할 때 교차검증 결과를 비교해서 선택"

## 교차검증용 통합 함수 cross_val_score
- 분류, 회귀 모두 사용 가능
- 타깃을 확인하고 적절한 폴드 방식을 선택.(SKfold,kfold)
- 사용할 머신러닝모델, 피쳐,레이블(target)을 전달합니다.
- 추가로 성능지표(ㅇㅇ예, accuracy), 폴드수도 선택 가능.

```python
from sklearn.model_selection import cross_val_score , cross_validate
from sklearn.datasets import load_iris
import numpy as np # numpy import 추가

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=1)

data = iris_data.data # 피처
label = iris_data.target # 레이블

# 성능 지표는 정확도(accuracy) , 교차 검증 세트는 3개
scores = cross_val_score(dt_clf , data , label , scoring='accuracy',cv=5)
#                    알고리즘 , 피쳐 ,레이블, 측정지표명,      폴드 수
print('교차 검증별 정확도:',np.round(scores, 4))
print('평균 검증 정확도:', np.round(np.mean(scores), 4))
```

## 알고리즘 후보 선정 및 교차검증
- 1. 결정트리
- 2. 랜덤포레스트
- 3. 그라디언트 부스팅
-  알고리즘 후보군 설정할 때는 챗봇의 도움을 받아도 됨

```python
# 모델 로드
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
dt_clf = DecisionTreeClassifier(random_state=1)
rf_clf = RandomForestClassifier(random_state=1)
gb_clf = GradientBoostingClassifier(random_state=1)
```

```python
# 알고리즘 후보군 전체 교차검증 진행
models = {'결정트리':dt_clf, '랜덤포레스트':rf_clf, '그레이디언트부스팅':gb_clf}

for name, model in models.items():
    scores = cross_val_score(model , data , label , scoring='accuracy',cv=5)


    print(f'모델명:{name}')
    print('교차 검증별 정확도:',np.round(scores, 4))
    print('평균 검증 정확도:', np.round(np.mean(scores), 4))
    print('-'*40)
```
